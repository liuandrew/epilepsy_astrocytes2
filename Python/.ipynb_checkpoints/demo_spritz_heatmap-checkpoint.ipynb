# Testing Spritz Classification Function

```python
def classify_peaks(t, c, number_of_stimulations=30):
    # analyze the calcium time course and tell me if it is 1:1, or 1:2
    
    peak_idxs = scipy.signal.find_peaks(c)[0] # where calcium spikes appear
    c_peaks = c[peak_idxs]
    number_of_responses = (c_peaks > 0.3).sum() # count up the number of calcium peaks larger than 300uM
    
    if number_of_responses < (number_of_stimulations-20):
        return 3
    if number_of_responses < (number_of_stimulations-2):
        return 2
    else:
        return 1
```

```python
run_spritz(period=19, dur=1, conc=10, num=30)
plot_experiment_plots(['glut', 'c'])

# Test the classification on this calcium time course
classify_peaks(cfg.t, cfg.c)
```

# Double for loop for spritzes

```python
periods = [10, 13, 16, 19, 22, 25, 28]
concentrations = [10, 20, 30, 40]

classifications = []

for period in periods:
    
    row = []
    for conc in concentrations:
        
        # For each period and concentration, run the experiment
        #  classify the behavior and add it to a row
        run_spritz(period=period, dur=1, conc=conc, num=30)
        behavior = classify_peaks(cfg.t, cfg.c)
        row.append(behavior)
        
        # print(period, conc, behavior)
        
    # Add the whole row to a bigger matrix to make a 2d array
    print(row)
    classifications.append(row)
```

```python
plt.imshow(classifications)
```
