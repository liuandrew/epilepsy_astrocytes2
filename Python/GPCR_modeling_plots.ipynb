# Equilibrium Gd1 and Gd2

For different glutamate concentrations, what are the equilibrium values of Gd1 and Gd2

```python
cfg.input_duration = 10000
gluts = np.linspace(0, 0.5, 300)
equil_Gd1s = []
equil_Gd2s = []
for glut in tqdm(gluts):
    cfg.input_max = glut
    run_gpcr_experiment('pulse', t_f=5000, max_step=10)
    equil_Gd1s.append(cfg.Gd1[-1])
    equil_Gd2s.append(cfg.Gd2[-1])
    # plot_experiment_plots(['glut', 'Gstar', 'Gd1', 'Gd2'])
    if np.abs(cfg.Gd1[-1] - cfg.Gd1[-2]) > 1e-5 or \
        np.abs(cfg.Gd2[-1] - cfg.Gd2[-2]) > 1e-5:
        print('problem')

pickle.dump([gluts, equil_Gd1s, equil_Gd2s], open('data/plot_data/equilibrium_GPCR', 'wb'))
```
