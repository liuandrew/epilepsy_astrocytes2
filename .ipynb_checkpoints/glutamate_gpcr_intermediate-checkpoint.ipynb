# GPCR ODE Model

This is the ODE system for the glutamate GPCR receptor inside of an astrocyte. The model is based on the following diagram:

<img src="GCPR_model.jpg" width=400 />

|Variable|Meaning|Python variable name|
|---|---|---|
|$G$|Inactivate glutamate receptor|G|
|$G^*$|Activated glutamate receptor|Gstar|
|$G_{d1}$|Deactivated (homologous) receptor|Gd1|
|$G_{d2}$|Deactivated (heterlogous) receptor|Gd2|
|$\lambda$|Downstream Gd2 Signal|lamb|

|Parameter|Meaning|Python variable name
|---|---|---|
|$k_+$|Activation constant $G \rightarrow G^*$|kp|
|$k_-$|Inactivation constant $G^* \rightarrow G$|km|
|$\gamma$|Concentration of glutamate|glut|
|$k_{d1}$|Deactivation (homologous) $G^* \rightarrow G_{d1}$|kd1|
|$k_{r1}$|Reactivation $G_{d1} \rightarrow G$|kr1|
|$k_{d2}$|Deactivation (heterlogous) $G \rightarrow G_{d2}$|kd2|
|$k_{r2}$|Reactivation $G_{d2} \rightarrow G$|kr2|
|$k_{\lambda+}$|$\lambda$ Generation term|klp|
|$k_{\lambda-}$|$\lambda$ Degradation term|klm

The ODE equations are as following:
$$ 
\begin{align}
% \frac{dG^*}{dt} &= k_+ \gamma G - k_- G^* - k_{d1} G^* \\
\frac{dG^*}{dt} &= k_+ \gamma G - k_- G^* - k_{d1} G^* - k_{d2}G^*\lambda \\
\frac{dG_{d1}}{dt} &= k_{d1}G^* - k_{r1} G_{d1} \\
% \frac{dG_{d2}}{dt} &= k_{d2}G^* G - k_{r2} G_{d2} \\
\frac{dG_{d2}}{dt} &= k_{d2}(G^*+G)\lambda - k_{r2} G_{d2} \\
\frac{d\lambda}{dt} &= k_{\lambda+}G^* - k_{\lambda-}\lambda
\end{align}
$$
and $G$ is given implicitly as $G = 1 - G^* - G_{d1} - G_{d2}$

```python
'''
Import necessary packages
'''
import numpy as np
import scipy
import scipy.integrate
import scipy.signal
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib
import os
import json
from tqdm import tqdm
import pandas as pd

#set figure font sizes for readability
font = {'size' : 30,
       'family': 'serif',
       'sans-serif': ['Helvetica']}
matplotlib.rc('font', **font)
matplotlib.rc('text', usetex=True)
```

```python
'''
Initiate parameter values
'''
#G <-> G*
kp = 0.03 #activation rate
km = 0.04 #inactivation rate

#G* -> Gd1 -> G
kd1 = 0.02 #0.01 #homologous (Gd1) deactivation rate
kr1 = 0.01 #0.005 #homologous reactivation rate
#G -> Gd2 -> G
# kd2 = 0.003 #0.0025 #heterologous (Gd2) deactivation rate
kd2 = 0.2
# kr2 = 0.0007 #0.0004 #heterologous reactivation rate
kr2 = 0.005

klp = 0.0002 #rate of lambda generation
klm = 0.004 #rate of lambda degradation

#Square wave glutamate pulse parameters
glut_start = 10
glut_duration = 750
glut_max = 10
glut_min = 0

#oscillation parameters
oscillation_on_duration = 50
oscillation_off_duration = 150
num_oscillations = 2

#Initial conditions
t_0 = 0
x_0 = [0, 0, 0, 1e-8]

#Final time
t_f = 1000



param_dict = {
    'kp': kp,
    'km': km,
    'kd1': kd1,
    'kr1': kr1,
    'kd2': kd2,
    'kr2': kr2,
    'klp': klp,
    'klm': klm
}

param_set_1 = {
    'kp': kp,
    'km': km,
    'kd1': 0.02,
    'kr1': 0.01,
    'kd2': 0.6,
    'kr2': 0.005,
    'klp': 0.0002,
    'klm': 0.0004
}

param_set_2 = {
    'kp': kp,
    'km': km,
    'kd1': 0.02,
    'kr1': 0.03,
    'kd2': 0.6,
    'kr2': 0.005,
    'klp': 0.0006,
    'klm': 0.012
}


#double exponential input parameters
compute_r_decay = lambda A, d_decay: -1 / d_decay * np.log(0.005 / A)
A = 0.2
d_rise = 21
r_rise = 0.002
d_decay = 97
r_decay = compute_r_decay(A, d_decay)
'''
Figure save location
'''
save = 'figures/glutamate_gpcr/'
```

```python
def ode_equations(t, x, input_type='pulse', gstar_to_gd2=True, sat_kinetics=False):
    '''
    ODE equations following the GPCR equations given earlier
    x[0] = Gstar
    x[1] = Gd1
    x[2] = Gd2
    x[3] = lambda
    This returns x_dot, which is an array of the right hand sides of 
    each of the ODE equations in the same order as above
    
    input_type: given by ODE args, can be 'pulse' or 'step'
    gstar_to_d2: whether to have the G* -> Gd2 arrow on or off (True/False)
    sat_kinetics: whether lambda should act by saturating kinetics
    '''
    global param_dict
    kp = param_dict['kp']
    km = param_dict['km']
    kd1 = param_dict['kd1']
    kd2 = param_dict['kd2']
    kr1 = param_dict['kr1']
    kr2 = param_dict['kr2']
    klp = param_dict['klp']
    klm = param_dict['klm']
    
    Gstar = x[0]
    Gd1 = x[1]
    Gd2 = x[2]
    lamb = x[3]
    G = 1 - Gstar - Gd1 - Gd2
    
    #query our glutamate function for what level glutamate is input
    if(input_type == 'pulse'):
        glut = glutamate_pulse_input(t)
    elif(input_type == 'step'):
        glut = glutamate_step_input(t)
    elif(input_type == 'oscillation'):
        glut = glutamate_oscillation_input(t)
    elif(input_type == 'exponential_oscillation'):
        glut = glutamate_exponential_oscillation(t)
    elif(input_type == 'none'):
        glut = 0
    
    x_dot = np.zeros(4)
    
    
    if(sat_kinetics):
        lamb_kinetics = lamb / (lamb + 0.2)
        klp = klp / 4
    else:
        lamb_kinetics = lamb
    
    x_dot[1] = kd1*Gstar - kr1*Gd1
    #This one has the G* -> Gd2 arrow on
    if(gstar_to_gd2):
        x_dot[0] = kp*glut*G - km*Gstar - kd1*Gstar - kd2*Gstar*lamb
        x_dot[2] = kd2*(Gstar+G)*lamb_kinetics - kr2*Gd2
    #This one has the G* -> Gd2 arrow OFF
    else:
        x_dot[0] = kp*glut*G - km*Gstar
        x_dot[2] = kd2*G*lamb_kinetics - kr2*Gd2
    
    #x_dot[2] = kd2*Gstar*G - kr2*Gd2
#     x_dot[2] = kd2*Gstar*G*(Gd2+0.01) - kr2*Gd2
#     x_dot[3] = klp*Gstar - klm*lamb + klp*0.01*lamb**2/(lamb**2 + 0.5)
    x_dot[3] = klp*Gstar - klm*lamb
    
    #These actually have some pretty interesting dynamic properties
#     x_dot[3] = klp*Gstar - klm*lamb + klm*11*lamb**3/(lamb**3 + 0.002)
#     x_dot[3] = klp*Gstar - klm*lamb + klm*0.5*lamb**2/(lamb**2 + 0.02)
    
    return x_dot



def glutamate_pulse_input(t):
    '''
    This function tells us what value glutamate takes at a given time t
    Currently set to square wave
    '''
    if(t > glut_start and t < (glut_start + glut_duration)):
        glut = glut_max
    else:
        glut = glut_min
    return glut


def glutamate_step_input(t):
    '''
    This function will create an IP3 input that increases stepwise
    0-50ms: 0
    50-200ms: 0.125
    200-350ms: 0.250
    350-500ms: 0.375
    500-650ms: 0.5
    650-800ms: 0.375
    800-950ms: 0.250
    950-1100ms: 0.125
    1100ms-: 0
    '''
    #times at which the concentration changes
    time_breaks = [0, 50, 200, 350, 500, 650, 800, 950, 1100]
    #values the concentration changes to at each interval
    ip3_values = [0, 0.5, 1, 1.5, 2, 1.5, 1, 0.5, 0]

    for i in range(len(time_breaks)):
        #check if t is after last time_break
        if(i == len(time_breaks) - 1):
            if(t >= time_breaks[i]):
                return ip3_values[i]

        #check which interval t is in
        else:
            if(t >= time_breaks[i] and t < time_breaks[i+1]):
                return ip3_values[i]
            
            
            
def glutamate_oscillation_input(t):
    '''
    This function creates an oscillatory glutamate input
    '''
    global num_oscillations
    global oscillation_on_duration
    global oscillation_off_duration
    
    principle_t = (t) % (oscillation_on_duration + oscillation_off_duration)
    
#     print((oscillation_off_duration + oscillation_on_duration) * num_oscillations)
    if(principle_t < oscillation_on_duration and
        t < (oscillation_off_duration + oscillation_on_duration) * num_oscillations):
        return glut_max
    else:
        return glut_min
    
    
    

def glutamate_exponential_oscillation(t):
    '''
    This function creates an oscillatory glutamate input where the input grows
    and falls exponentially
    '''
    #compute our IP3 curve helper functions
    oscillation_on_duration = d_rise + d_decay
    principle_t = t % (oscillation_on_duration + oscillation_off_duration)
    
    
    if(principle_t < oscillation_on_duration and
      t < (oscillation_on_duration + oscillation_off_duration) * num_oscillations):
        s_inf = A / (1 - np.exp(-r_rise * d_rise))
        if(principle_t < d_rise):
            return s_inf * (1 - np.exp(-r_rise * (principle_t)))
        elif(principle_t >= d_rise):
            return A * np.exp(-r_decay * (principle_t - d_rise))
    else:
        return 0
    

            
def get_input_plot(input_type, t_f=1000):
    '''
    This function simply returns t and y values to plot
    based on the type of input curve that we used
    E.g., we could call this to get glut_t, glut for the glutamate
    input plot
    '''
    t = np.arange(t_0, t_f, t_f/100000)
    y = []
    for x in t:
        if(input_type == 'pulse'):
            y.append(glutamate_pulse_input(x))
        elif(input_type == 'step'):
            y.append(glutamate_step_input(x))
        elif(input_type == 'oscillation'):
            y.append(glutamate_oscillation_input(x))
        elif(input_type == 'curve'):
            y.append(ip3_curve_input(x))
        elif(input_type == 'exponential_oscillation'):
            y.append(glutamate_exponential_oscillation(x))
        elif(input_type == 'none'):
            y.append(0)
    y = np.array(y)
    return t, y


def set_gpcr_params(set_num):
    '''
    Set params to be either the primary (set_num=1) or secondary (set_num=2)
    '''
#     global param_set_1
#     global param_set_2
    if(set_num == 1):
        params = param_set_1
    elif(set_num == 2):
        params = param_set_2
    else:
        raise Exception('No valid value given for set_num. Must pass either 1 or 2')
        
    for param in params:
        param_dict[param] = params[param]
```

```python
def run_experiment(change_param_dict={}, t_0=0, t_f=1000, max_step=1, input_type='pulse',
                  make_plot=True, save_plot=False, gstar_to_gd2=True, sat_kinetics=False):
    '''
    Run the ode simulation with the given change_param_dict values
    Note that we will revert the param dict back to default settings after running the simulation
    
    t_0: start time
    t_f: end time
    input_type: 'pulse'/'step'/'oscillation'
    make_plot: whether to create the plot
    save_plot: whether to save the plot. A string should be passed for the name
        e.g. if 'test' is passed, we will save 'figures/glutamate_gpcr/test.pdf'
    gstar_to_gd2: whether to turn the G* -> Gd2 arrow on or off
    sat_kinetics: whether lambda should act by saturating kinetics
    '''
    global param_dict
    global kp
    global km
    global kd1
    global kd2
    global kr1
    global kr2
    global klp
    global klm
    
    global Gstar
    global Gd1
    global Gd2
    global G
    global lamb
    global t
    for key in change_param_dict.keys():
        param_dict[key] = change_param_dict[key]
    
    #run the numerical ODE solver
    sol = scipy.integrate.solve_ivp(ode_equations, [t_0, t_f], x_0, 
                                    args=[input_type, gstar_to_gd2, sat_kinetics], max_step=max_step)

    #get the output arrays from the solver
    Gstar = sol.y[0]
    Gd1 = sol.y[1]
    Gd2 = sol.y[2]
    lamb = sol.y[3]
    G = np.full(len(Gstar), 1) - Gstar - Gd1 - Gd2
    t = sol.t
    
    sol['glutamate_input_type'] = input_type
    
    #if make_plot == True, make the plot
    if(make_plot):
        plt.figure(figsize=(10,20))

        fig, ax = plt.subplots(6, 1, sharex=True, figsize=(10,10))

        #set glutamate plot curve
        glut_t, glut = get_input_plot(sol['glutamate_input_type'], t_f=t_f)

        #plot figures
        ax[0].plot(glut_t, glut)
        ax[1].plot(t, Gstar)
        ax[2].plot(t, G)
        ax[3].plot(t, lamb)
        ax[4].plot(t, Gd1)
        ax[5].plot(t, Gd2)

        #set xlabel
        ax[5].set_xlabel('t')

        #set ylabels
        ax[0].set_ylabel(r'$\phi$', rotation='horizontal', ha='left')
        ax[1].set_ylabel(r'$G^*$', rotation='horizontal', ha='left')
        ax[2].set_ylabel(r'$G$', rotation='horizontal', ha='left')
        ax[3].set_ylabel(r'$\lambda$', rotation='horizontal', ha='left')
        ax[4].set_ylabel(r'$G_{d1}$', rotation='horizontal', ha='left')
        ax[5].set_ylabel(r'$G_{d2}$', rotation='horizontal', ha='left')

        #align ylabels
        ax[0].get_yaxis().set_label_coords(-0.23,0.4)
        ax[1].get_yaxis().set_label_coords(-0.23,0.4)
        ax[2].get_yaxis().set_label_coords(-0.23,0.4)
        ax[3].get_yaxis().set_label_coords(-0.23,0.35)
        ax[4].get_yaxis().set_label_coords(-0.23,0.35)
        ax[5].get_yaxis().set_label_coords(-0.23,0.35)

        fig.tight_layout()
        
    #save the figure as the given name
    if(save_plot != False and make_plot):
        if('figures' not in os.listdir()):
            os.mkdir('figures')
        if('glutamate_gpcr' not in os.listdir('figures')):
            os.mkdir('figures/glutamate_gpcr')
        
        fig.savefig('figures/glutamate_gpcr/' + save_plot + '.png', bbox_inches='tight')
        
        
    #reset the global param_dict back to default values after experiment
    param_dict = {
        'km': km,
        'kp': kp,
        'kd1': kd1,
        'kd2': kd2,
        'kr1': kr1,
        'kr2': kr2,
        'klp': klp,
        'klm': klm
    }
    
    #return the solution found
    return sol
```

```python
'''
Functions used to plot out parameter search results (all used in section 5)
'''

param_results = pd.read_pickle('./data/oscillation_param_peaks')
data = pd.read_pickle('./data/gpcr_parameter_search_oscillation')

def get_experiments_with_params(row):
    '''
    Pass in a row index for search_results dataframe to get
    the rows from the original experiment with the given parameter set
    '''
    row = search_results.loc[row]
    experiments = data[(data['kd1'] == row['kd1']) & (data['kd2'] == row['kd2']) & 
                        (data['kr1'] == row['kr1']) & (data['kr2'] == row['kr2']) & 
                        (data['klp'] == row['klp']) & (data['klm'] == row['klm'])]    
    return experiments




def plot_experiments(row, oscillation_off_duration=None, glut=None):
    '''
    Plot experiments with param set from the given row index of search_results
    Plot with a fixed oscillation_off_duration or glutamate level
    '''
    experiments = get_experiments_with_params(row)
    if(oscillation_off_duration is not None):
        #Oscillation off duration specified, plot different glut levels
        experiments = experiments[experiments['oscillation_off_duration'] == oscillation_off_duration]
        fig, ax = plt.subplots(6, 4, figsize=(16, 12))
        for n, i in enumerate(experiments.index):
            read_row(i)
            add_xlabel = False
            add_ylabels = False
            remove_yticks = True
            if(n == 0):
                add_ylabels = True
                add_xlabel = True
                remove_yticks = False
                
            plot_experiment_plots(['glut', 'Gstar', 'G', 'lamb', 'Gd1', 'Gd2'], ax[:, n],
                                 plot_input=True, input_type='oscillation', t_f=2000,
                                 add_xlabel=add_xlabel, add_ylabels=add_ylabels, remove_yticks=remove_yticks)
            ax[0, n].set_title(r'$\phi=$' + str(glut_max))
            
    elif(glut is not None):
        experiments = experiments[experiments['glut_max'] == glut]
        fig, ax = plt.subplots(6, 3, figsize=(16, 12))
        for n, i in enumerate(experiments.index):
            read_row(i)
            add_xlabel = False
            add_ylabels = False
            remove_yticks = True
            if(n == 0):
                add_ylabels = True
                add_xlabel = True
                remove_yticks = False
                
            plot_experiment_plots(['glut', 'Gstar', 'G', 'lamb', 'Gd1', 'Gd2'], ax[:, n],
                                 plot_input=True, input_type='oscillation', t_f=2000,
                                 add_xlabel=add_xlabel, add_ylabels=add_ylabels, remove_yticks=remove_yticks,
                                 ylabel_padding=[-0.5, 0.4])
            ax[0, n].set_title(r'$\phi=$' + str(glut_max))
            
            
            
            
def read_row(i):
    '''
    Load the data from a row of data data_frame (which holds experiment results)
    into global data so we can plot the data
    '''
    global y
    global t
    global t_input
    global Gstar
    global Gd1
    global Gd2
    global G
    global lamb
    global oscillation_on_duration
    global oscillation_off_duration
    global glut_max
    
    y = np.frombuffer(data.loc[i, 'y']).reshape(data.loc[i, 'y_shape'])
    t = np.frombuffer(data.loc[i, 't'])

    Gstar = y[0]
    Gd1 = y[1]
    Gd2 = y[2]
    lamb = y[3]
    G = np.full(len(t), 1) - Gstar - Gd1 - Gd2
    
    oscillation_on_duration = data.loc[i, 'oscillation_on_duration']
    oscillation_off_duration = data.loc[i, 'oscillation_off_duration']
    glut_max = data.loc[i, 'glut_max']  
    
    
    
    
def plot_experiment_plots(variables, axs, add_ylabels=True, add_xlabel=True, plot_input=True, 
                          ylabel_padding=[-0.23, 0.4], input_type='pulse', t_f=1000, remove_yticks=False):
    '''
    Plot the solutions of the numerical solver for multiple variables
    Use the passed axs, iterating through them one by one and plotting the variables in the given order
    
    variables: list of variable names e.g. ['Gstar', 'p', 'h', 'c']
    axs: list of axes to plot on. For example a column of plots from plt.subplots
    add_ylabels: whether to add ylabels to the plot
    add_xlabel: whether to add xlabel of time to the last axis
    plot_input: whether first plot is an input (then we will use t_input instead of t)
    ylabel_padding: padding for each ylabel (first is x direction, second is y direction)
    
    This should only be used if more than one variable is being plotted

    Ex.
    run_Gstar_controlled_experiment('pulse')
    fig, ax = plt.subplots(4, 1, figsize=(10, 5))
    plot_experiment_plots(['Gstar', 'p', 'h', 'c'], ax)
    '''
    ylabels = {
        'glut': r'$\phi$',
        'Gstar': r'$G^*$',
        'Gd1': r'Gd1',
        'Gd2': r'Gd2',
        'G': r'G',
        'lamb': r'$\lambda$',
        'p': r'IP$_3$',
        'h': r'$h$',
        'c': r'$c$',
        'c_tot': r'$c_{tot}$'
    }
    
    for i, variable in enumerate(variables):
        y = globals()[variable]
        
        if(plot_input and i==0):
#             axs[i].plot(t_input, y)
            t_input, y = get_input_plot(input_type, t_f=t_f)
            axs[i].plot(t_input, y)
        else:
            axs[i].plot(t, y)
    
    if(add_ylabels):
        for i, variable in enumerate(variables):
            axs[i].set_ylabel(ylabels[variable], rotation='horizontal', ha='left')
            axs[i].get_yaxis().set_label_coords(ylabel_padding[0], ylabel_padding[1])        
            
    if(add_xlabel):
        axs[len(variables) - 1].set_xlabel(r'$t$')
        
    if(remove_yticks):
        for i, variables in enumerate(variables):
            axs[i].set_yticks([])
```

# Square Waves
Here we can get "slow" growth of Gd2 over square wave stimulation, while Gd1 increases and decreases over the period of the oscillation off duration. A few things of note
* It is fairly easy to manipulate the characteristic timescale of Gd2/$\lambda$ dynamics, by simply multiplying kd2&kr2 or klp&klm by the same constant. This changes the timescale without affecting the steady state values
* Gd2 noticeably grows even during the off period, which may be desirable, but this can be corrected by increasing the timescale of $\lambda$
    * Increasing the timescale of $\lambda$ simultaneously causes Gd2 to increase to its maximum more quickly, which may not be desirable
    
!
* With klm = 0.001, kd2 = 0.1, we have a gradual enough growth of Gd2 so that the second and third peaks are distinguishible. However, the Gd2 degradation takes way too long (our goal is to have most Gd2 recovered by 10 minutes of off stimulation)
* On the other hand, if klm = 0.004, Gd2 degradation is roughly correct, but does not build up strong enough to be very noticeable in desensitization. If kd2 = 0.4, it is noticeable but now builds up almost completely by the first peak

# Experimental Suite
We would like to run the following experiments with a range of parameters
1. 30 second double bath application with 2/4/10 minute off periods
    1. Clear desensitization at 2 minutes
    2. Mostly recovered at 4 minutes
    3. Near full recovery at 10 minutes
2. 60ms pulses with 15 seconds in between - should have largely the same response at each pulse

The following code performs the experiments for a range of parameters. Basically we have a base parameter set for $k_{r1}$, $k_{d1}$, $k_{r2}$, $k_{d2}$, $k_{\lambda+}$, $k_{\lambda-}$, and then increase or decrease each for 3x, thus we have $3^6$ parameter sets

```python
data = pd.DataFrame(columns=['input_type', 'oscillation_on_duration', 'oscillation_off_duration',
                     'max_step', 'glut_max', 'kd1', 'kr1', 'kd2', 'kr2', 'klp', 'klm', 'y', 'y_shape', 't'])

#Experimental conditions, 1A, 1B, 1C, 2
# input_types = ['oscillation', 'oscillation', 'oscillation', 'oscillation']
# oscillation_on_durations = [30, 30, 30, 0.06]
# max_steps = [1, 1, 1, 0.02]
# oscillation_off_durations = [120, 240, 600, 15]

#Skipping experimental condition 2 since it is way slower
oscillation_on_durations = [30, 30, 30]
oscillation_off_durations = [120, 240, 600]


#Glutamate strengths
glut_maxes = [0.1, 0.2, 0.5, 1]

#Parameter multipliers
#Each parameter will be multiplied by the constant from its current value
#"Base" values: kd1=0.02, kr1=0.01, kd2=0.2, kr2=0.005, klp=0.0002, klm=0.004
# multipliers = np.array([0.125, 0.25, 0.5, 1, 2, 4, 8])
multipliers = np.array([0.3, 1, 3])
# multipliers = np.array([1.])
kd1_grid, kr1_grid, kd2_grid, kr2_grid, klp_grid, klm_grid = \
    np.meshgrid(kd1*multipliers, kr1*multipliers, kd2*multipliers,
                kr2*multipliers, klp*multipliers, klm*multipliers)

trial = 0
for i in tqdm(range(len(oscillation_on_durations)), desc='outermost'):
    for glut_max in tqdm(glut_maxes, desc='second'):
        for kd1_temp in kd1*multipliers:
            for kr1_temp in kr1*multipliers:
                for kd2_temp in kd2*multipliers:
                    for kr2_temp in kr2*multipliers:
                        for klp_temp in klp*multipliers:
                            for klm_temp in klm*multipliers:
                                oscillation_on_duration = oscillation_on_durations[i]
                                oscillation_off_duration = oscillation_off_durations[i]
                                sol = run_experiment({'kd1': kd1_temp, 'kr1': kr1_temp, 'kd2': kd2_temp,
                                                     'kr2': kr2_temp, 'klp': klp_temp, 'klm': klm_temp},
                                                    make_plot=False, sat_kinetics=False, t_f=2000, input_type='oscillation')
                                results_dict = {
                                    'input_type': 'oscillation',
                                    'oscillation_on_duration': oscillation_on_duration,
                                    'oscillation_off_duration': oscillation_off_duration,
                                    'max_step': 1,
                                    'glut_max': glut_max,
                                    'kd1': kd1_temp,
                                    'kr1': kr1_temp,
                                    'kd2': kd2_temp,
                                    'kr2': kr2_temp,
                                    'klp': klp_temp,
                                    'klm': klm_temp,
                                    'y': sol.y.tostring(),
                                    't': sol.t.tostring(),
                                    'y_shape': sol.y.shape
                                }
                                data.loc[trial] = results_dict

#                                 if(trial % 100 == 9):
#                                     print(trial)
                                trial += 1
                                
        
data.to_pickle('./data/gpcr_parameter_search_oscillation')
```

# Read the data and classify different parameter regimes

After running all of the experiments, we first determine the ratio of signal response between the second glutamate square pulse and the first. Basically we determine how much desensitization there is for different parameter sets.

This code organizes the experiments from the 'data' dataframe and summarizes it into these signal ratios in the 'param_results' dataframe. Finally we filter through the parameter sets into 'search_results' to look for a parameter set with desirable properties.

```python
#This is the exact code inside the compute_response_differences function below
#We can plug in a specific row from the data data_frame (which holds simulation results)
#and see what the signal processor reads
#Peak: ratio of heights of peak2:peak1
#Integral: ratio of total signal between signal2:signal1, as delimited by circles on the plot


read_row(800)

#Get response times as the start of glutamate oscillation turning on
start_times = [t*(oscillation_on_duration + oscillation_off_duration) for t in range(2)]
start_indices = [np.argmax(t > T) for T in start_times]

#Get peaks and find when the final response ends
peaks = scipy.signal.find_peaks(np.array(Gstar))[0]
final_peak = peaks[-1]
end_response = np.argmax(Gstar[final_peak:] < 1e-4) + final_peak
Gstar_peaks = Gstar[peaks]

#Integrate signal responses
total_responses = []
response_indices = start_indices + [end_response]
for i in range(len(response_indices) - 1):
    response_start = response_indices[i]
    response_end = response_indices[i + 1]
    delta_t = np.diff(t[response_start:response_end]) #the widths of each time step
    response_heights = Gstar[response_start:response_end - 1]
    total_response = np.sum(delta_t * response_heights)
    total_responses.append(total_response)
    
#Compute ratio differences in peak and total response heights
peak_ratios = Gstar[peaks[1]] / Gstar[peaks[0]]
integral_ratios = total_responses[1] / total_responses[0]

print(f'Peak: {peak_ratios} \t Integral: {integral_ratios}')

plt.plot(t, Gstar)
plt.scatter(t[start_indices], Gstar[start_indices])
plt.scatter(t[end_response], Gstar[end_response])
```

```python
'''
This block of code calculates the response and peak ratios for every experiment that we run
'''

def compute_response_differences(row, plot=False):
    '''
    Compute the response characterstic differences for different rows simulated
    '''
    read_row(row)

    #Get response times as the start of glutamate oscillation turning on
    start_times = [t*(oscillation_on_duration + oscillation_off_duration) for t in range(2)]
    start_indices = [np.argmax(t > T) for T in start_times]

    #Get peaks and find when the final response ends
    peaks = scipy.signal.find_peaks(np.array(Gstar))[0]
    final_peak = peaks[-1]
    end_response = np.argmax(Gstar[final_peak:] < 1e-4) + final_peak
    Gstar_peaks = Gstar[peaks]

    #Integrate signal responses
    total_responses = []
    response_indices = start_indices + [end_response]
    for i in range(len(response_indices) - 1):
        response_start = response_indices[i]
        response_end = response_indices[i + 1]
        delta_t = np.diff(t[response_start:response_end]) #the widths of each time step
        response_heights = Gstar[response_start:response_end - 1]
        total_response = np.sum(delta_t * response_heights)
        total_responses.append(total_response)

    #Compute ratio differences in peak and total response heights
    peak_ratios = Gstar[peaks[1]] / Gstar[peaks[0]]
    integral_ratios = total_responses[1] / total_responses[0]

#     print(f'Peak: {peak_ratios} \t Integral: {integral_ratios}')

    if(plot):
        plt.plot(t, Gstar)
        plt.scatter(t[start_indices], Gstar[start_indices])
        plt.scatter(t[end_response], Gstar[end_response])
        
    return peak_ratios, integral_ratios

for i in tqdm(range(len(data))):
    peak_ratios, integral_ratios = compute_response_differences(i)
    data.loc[i, 'peak_ratios'] = peak_ratios
    data.loc[i, 'integral_ratios'] = integral_ratios
```

```python
'''
Test the data to see which sets of parameters fit the requirements
This code converts every ODE simulation that we run into a table called param_results
    which is organized by parameter set and summarizes the different between two peak heights
    for each glutamate level and time between the two square pulses
Data is written to param_results
'''

columns = ['kd1', 'kr1', 'kd2', 'kr2', 'klp', 'klm']
for i in range(3):
    oscillation_on_duration = oscillation_on_durations[i]
    oscillation_off_duration = oscillation_off_durations[i]
    for glut_max in glut_maxes:
        columns += [f'{oscillation_on_duration}on{oscillation_off_duration}off{float(glut_max)}glut']
        
param_results = pd.DataFrame(columns=columns)

oscillation_on_durations = [30, 30, 30]
oscillation_off_durations = [120, 240, 600]
glut_maxes = [0.1, 0.2, 0.5, 1.0]

multipliers = np.array([0.3, 1, 3])

for kd1_temp in kd1*multipliers:
    for kr1_temp in kr1*multipliers:
        for kd2_temp in kd2*multipliers:
            for kr2_temp in kr2*multipliers:
                for klp_temp in klp*multipliers:
                    for klm_temp in klm*multipliers:
                        results = data[(data['kd1'] == kd1_temp) & (data['kr1'] == kr1_temp) &
                                       (data['kd2'] == kd2_temp) & (data['kr2'] == kr2_temp) &
                                       (data['klp'] == klp_temp) & (data['klm'] == klm_temp)]
                        results_dict = {
                            'kd1': kd1_temp,
                            'kr1': kr1_temp,
                            'kd2': kd2_temp,
                            'kr2': kr2_temp,
                            'klp': klp_temp,
                            'klm': klm_temp,
                        }

                        
                        for i in range(3):
                            oscillation_on_duration = oscillation_on_durations[i]
                            oscillation_off_duration = oscillation_off_durations[i]
                            for glut_max in glut_maxes:
                                experiment_index = results[(results['oscillation_on_duration'] == oscillation_on_duration) &
                                                          (results['oscillation_off_duration'] == oscillation_off_duration) & 
                                                          (results['glut_max'] == glut_max)].index[0]
                                ratios, _ = compute_response_differences(experiment_index)
                                results_dict[f'{oscillation_on_duration}on{oscillation_off_duration}off{glut_max}glut'] = ratios
                        param_results.loc[len(param_results)] = results_dict
                        
# param_results.to_pickle('./data/oscillation_param_peaks')
```
